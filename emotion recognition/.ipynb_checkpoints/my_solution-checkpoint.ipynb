```python
train_data = pd.read_csv('train.csv',index_col=0)
train_data['comment_text'] = train_data['comment_text'].astype(str)
```

## __Using BERT transformer for recognition__

```python
from transformers import BertTokenizer, BertConfig,AdamW, BertForSequenceClassification,get_linear_schedule_with_warmup
```

```python
MAX_LEN = 256 #max len in article is 512 !!!!!!
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased',do_lower_case=True) 
input_ids = [tokenizer.encode(tmp, add_special_tokens=True,max_length=MAX_LEN,pad_to_max_length=True,truncation=True)
             for tmp in train_data['comment_text']]

## attention mask
attention_masks = []
attention_masks = [[float(i>0) for i in seq] for seq in input_ids]
```

### Train_valid split

```python
X_train, X_val, y_train, y_val = train_test_split(input_ids, train_data['toxicity'].values, 
                                                  test_size=0.1, random_state=0, stratify=train_data['toxicity'])
X_train = torch.tensor(X_train)
X_val = torch.tensor(X_val)
y_train = torch.tensor(y_train)
y_val = torch.tensor(y_val)

masks_train, masks_val, y_masks_train, y_masks_val = train_test_split(attention_masks, 
                                                                      train_data['toxicity'], test_size=0.1, 
                                                                      random_state=0, stratify=train_data['toxicity'])
masks_train = torch.tensor(masks_train)
masks_val = torch.tensor(masks_val)
```

```python
from torch.utils.data import TensorDataset, DataLoader, RandomSampler, SequentialSampler
```

```python
batch_size = 32
```

```python
train = TensorDataset(X_train,masks_train,y_train) 
train_sampler = RandomSampler(train)
train_dataloader = DataLoader(train,sampler=train_sampler,batch_size=batch_size)

valid = TensorDataset(X_val,masks_val,y_val)
valid_sampler = RandomSampler(valid)
valid_dataloader = DataLoader(valid,sampler=valid_sampler,batch_size=batch_size)
```

### Pre-trained model initialization

```python
model = BertForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=6).to(device)

lr = 2e-5
adam_epsilon = 1e-8
epochs = 5

num_warmup_steps = 0
num_training_steps = len(train_dataloader)*epochs

optimizer = AdamW(model.parameters(), lr=lr,eps=adam_epsilon,correct_bias=False)  # To reproduce BertAdam specific behavior set correct_bias=False
scheduler = get_linear_schedule_with_warmup(optimizer, num_warmup_steps=num_warmup_steps, num_training_steps=num_training_steps)  # PyTorch scheduler
```

### FItting Model

#### Validation

```python
accuracy = []
learning_rate = []

model.zero_grad() #MUST HAVE

for i in tnrange(0,epochs,desc='Epoch'):
    
    print(F" Epoch {i} ")

    batch_loss = 0
    train_dataloader = tqdm(train_dataloader)
    for step, batch in enumerate(train_dataloader):
        #Initialization
        model.train() #set up training mod
        batch = tuple(t.to(device) for t in batch) #add to the GPU (if it available)
        b_input_ids, b_input_mask, b_labels = batch #unpacking
        
        #Foraward propagation
        outputs = model(b_input_ids, token_type_ids=None, 
                        attention_mask=b_input_mask, 
                        labels=b_labels)
        loss = outputs[0]
        
        #Backward propagation
        loss.backward()
        
        #Update
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0) #clip to avoid exploding gradients
        optimizer.step() #update parameters and embeddings
        scheduler.step() #update learning rate schedule
        optimizer.zero_grad() #clear previous gradients
        batch_loss += loss.item() 
    
    
    avg_train_loss = batch_loss / len(train_dataloader) #current loss
    accuracy.append(avg_train_loss)
    print(F'\n\tAverage Training loss: {avg_train_loss}')
    
    for param_group in optimizer.param_groups: #current learning_rate
        print("\n\tCurrent Learning rate: ",param_group['lr'])
        learning_rate.append(param_group['lr'])
    
    #Validation    
    model.eval() #set up validation mod
    eval_accuracy, eval_mcc_accuracy, nb_eval_steps = 0, 0, 0
    
    for batch in valid_dataloader:
        batch = tuple(t.to(device) for t in batch) #add to the GPU (if it available)
        b_input_ids, b_input_mask, b_labels = batch  #unpacking
        
        #predictions
        with torch.no_grad():
            logits = model(b_input_ids, token_type_ids=None, attention_mask=b_input_mask)
              
        logits = logits[0].to('cpu').numpy() #add to the CPU (if GPU available)
        label_ids = b_labels.to('cpu').numpy() #add to the CPU (if GPU available)
        pred_flat = np.argmax(logits, axis=1).flatten()
        labels_flat = label_ids.flatten()
    
        df_metrics=pd.DataFrame({'Epoch':epochs,'Actual_class':labels_flat,'Predicted_class':pred_flat})

        tmp_eval_accuracy = accuracy_score(labels_flat,pred_flat)
        tmp_eval_mcc_accuracy = matthews_corrcoef(labels_flat, pred_flat)

        eval_accuracy += tmp_eval_accuracy
        eval_mcc_accuracy += tmp_eval_mcc_accuracy
        nb_eval_steps += 1

    print(F'\n\tValidation Accuracy: {eval_accuracy/nb_eval_steps}')
    print(F'\n\tValidation MCC Accuracy: {eval_mcc_accuracy/nb_eval_steps}')
```
